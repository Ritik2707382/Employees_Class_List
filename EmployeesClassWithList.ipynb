```python
class Employees:
  def __init__(self):
    self.employee=[]
  def add_e(self,name,salary,age,position,id):
    self.employee.append({
    'name':name,
    'salary':salary,
    'age':age,
    'position':position,
    'id':id
    })
    print(f"New Employees are added successfully")
  def remove_e_by_id(self,id):
    for emp in self.employee:
      if emp['id']==id:
        self.employee.remove(emp)
        print(f"Employee with Id {id} is removed successfully")
      else:
        print("Employee with Id doesn't exist")
  def remove_e_by_name(self,name):
    for emp in self.employee:
      if emp['name']==name:
        self.employee.remove(emp)
        print(f"Employee with Name {name} is removed successfully")
      else:
        print("Employee with Name doesn't exist")
  def update_by_id(self,id,name=None,salary=None,age=None,position=None):
    for emp in self.employee:
      if emp['id']==id:
        if name:
          emp['name']=name
        if salary:
          emp['salary']=salary
        if age:
          emp['age']=age
        if position:
          emp['position']=position
        print(f"Employee with Id {id} is updated successfully")
  def update_by_name(self,old_name,new_name=None,salary=None,age=None,position=None,id=None):
    for emp in self.employee:
      if emp['name']==old_name:
        if new_name:
          emp['new_name']=new_name
        if salary:
          emp['salary']=salary
        if age:
          emp['age']=age
        if position:
          emp['positon']=position
        if id:
          emp['id']=id
        print(f"Employee with Name {new_name} is updated successfully")
  def display_employee(self):
    if self.employee:
      for emp in self.employee:
        print(f"Name:{emp['name']} , Salary:{emp['salary']},  Age:{emp['age']},  Position:{emp['position']},  Id:{emp['id']}")
    else:
      print("No Elements to Display in Employee Class")
m1=Employees()
m2=Employees()
while True:
  print("\n 1.Add Employee" "\n 2.Remove Employee By Id" "\n 3.Remove Employee By Name" "\n 4.Update Employee By Id"
        "\n 5.Update Employee By Name" "\n 6.Display Employee Details" "\n 7.Exit")
  choice=int(input("Enter the Choice:"))
  if choice==1:
    name=input("Enter the Name:")
    salary=float(input("Enter the Salary:"))
    age=int(input("Enter the Age:"))
    position=input("Enter the Position:")
    id=int(input("Enter the Id:"))
    m1.add_e(name,salary,age,position,id)
  elif choice==2:
    id=int(input("Enter the ID to be Removed:"))
    m1.remove_e_by_id(id)
  elif choice==3:
    name=input("Enter the Name to be Removed:")
    m1.remove_e_by_name(name)
  elif choice==4:
    id=int(input("Enter the Id to be Updated:"))
    name=input("Enter the Name (Press Enter to Skip) to be Updated:")
    salary=input("Enter the Salary (Press Enter to Skip) to be Updated:")
    salary=float(salary) if salary else None
    age=input("Enter the Age(Press Enter to Skip) to be Updated:")
    age=int(age) if age else None
    position=input("Enter the Position(Press Enter to Skip)to be Updated:")
    m1.update_by_id(id,name,salary,age,position)
  elif choice==5:
    old_name=input("Enter the Current Name of Employee to be Updated:")
    new_name=input("Enter the New Name(Press Enter to Skip) to be Updated:")
    salary=input("Enter the Salary (Press Enter to Skip) to be Updated:")
    salary=float(salary) if salary else None
    age=input("Enter the Age(Press Enter to Skip) to be Updated:")
    age=int(age) if age else None
    position=input("Enter the Position(Press Enter to Skip)to be Updated:")
    m1.update_by_name(old_name,new_name,salary,age,position)
  elif choice==6:
    m1.display_employee()
  elif choice==7:
    print("Exit the Program")
    break
  else:
    print("Wrong Choice- Enter the Valid Choice")
```
